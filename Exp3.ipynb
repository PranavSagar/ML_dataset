```python
# # Data Preprocessing
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset= pd.read_csv('/content/drive/My Drive/Sem6th/ML/Exp3/Social_Network_Ads.csv')
```

```python
print(dataset)
```

```python
#Check Missing Data
dataset.isnull().sum()
```

```python
# Extracting dependent and independent variables:
# Extracting independent variable:
X = dataset.iloc[:, [2, 3]].values
#X = dataset.iloc[:, 1: 4].values
# Extracting dependent variable:
y = dataset.iloc[:, 4].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder  
label_encoder_x= LabelEncoder()  
X[:, 0]= label_encoder_x.fit_transform(X[:, 0])
print(X)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 5)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
print(y_train)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting Decision Tree Classification to the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
classifier.fit(X_train, y_train)
classifier.score(X_train,y_train)

#accuracy on training data
```

```python
#accuracy on test data
classifier.score(X_test,y_test)
```

```python
classifier.predict(X_test)
```

```python
#prob_a predicts probability of a sample belonging to class
#col 0 denotes proability of class 0 and col 1 denotes probabilty of class 1
classifier.predict_proba(X_test)
```

```python

y_pred=classifier.predict_proba(X_test)[:,1]
print(y_pred)
new_y=[]
for i in range (len(y_pred)):
  if y_pred[i]< 0.6:
    new_y.append(0)
  else:
    new_y.append(1)
print(new_y)

```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,new_y)
```

```python
#changing max depth (longest path of the tree)
train_accuracy=[]
valid_accuracy=[]
for depth in range (1,10):
  classifier=DecisionTreeClassifier(max_depth=depth, random_state=10)
  classifier.fit(X_train, y_train)
  train_accuracy.append(classifier.score(X_train,y_train))
  valid_accuracy.append(classifier.score(X_test,y_test))
```

```python
print(train_accuracy)
print(valid_accuracy)
```

```python
 classifier=DecisionTreeClassifier(max_depth=2, random_state=10)
 classifier.fit(X_train, y_train)
 classifier.score(X_train,y_train)
```

```python
classifier.score(X_test,y_test)
```

```python
frame=pd.DataFrame({'max_depth':range(1,10), 'train_acc':train_accuracy, 'test_acc':valid_accuracy})
frame.head(9)
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'], frame['train_acc'], marker='o')
plt.plot(frame['max_depth'], frame['test_acc'], marker='o')
plt.xlabel('depth of tree')
plt.ylabel('performance')
plt.legend()
```

```python
classifier=DecisionTreeClassifier(max_depth=2, random_state=10)
classifier.fit(X_train, y_train)
classifier.score(X_train,y_train)
```

```python
 classifier.score(X_test,y_test)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix: ')
print(cm)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy: %.2f' % (accuracy_score(y_test, y_pred)*100))
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
# calculate precision
# Precision = TruePositives / (TruePositives + FalsePositives) 
# calculate Accuracy
#   0   1
# 0[59  7(FP type 1)]
#1 [ 11(FN Type 2) 23]= 82/100
precision = precision_score(y_test, y_pred, average='binary')
print('Precision: %.3f' % (precision))
```

```python
# calculate recall
# Recall = TruePositives / (TruePositives + FalseNegatives)
# calculate Accuracy
#   0   1
# 0[59  7(FP type 1)]
#1 [ 11(FN Type 2) 23]= 82/100
recall = recall_score(y_test, y_pred, average='binary')
print('Recall: %.3f' % (recall))
```

```python
# F-Measure = (2 * Precision * Recall) / (Precision + Recall)
# (2*0.829*0.906)/0.829+0.906
# calculate score
score = f1_score(y_test, y_pred, average='binary')
print('F-Measure: %.3f' % (score*100))
```
